# SHALLOW NEURAL NETWORK


### Model Building

Importing Libraries

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import SGD             # Gradient descent (with momentum) optimizer : UPDATING WEIGHTRS WHILE BACK PROPAGATION
```

Splitting the Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
y_train.shape
```

Defining the Model

```python
model = Sequential()

# INPUT LAYER
model.add(Dense(13, input_dim=13, activation='relu'))

# HIDDEN LAYER
model.add(Dense(13, activation='relu'))

# OUTPUT LAYER
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=SGD(learning_rate=0.01), metrics=['accuracy'])
```


### Model Training

```python
model.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

loss, accuracy = model.evaluate(X_test, y_test)

print(f"\n\nTest loss: {loss}, Test accuracy: {accuracy}")
```

Predicting for test data

```python
y_pred = model.predict(X_test)
```


### Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,  f1_score, precision_score

print("ACCURACY : ", accuracy_score(y_test.astype("int32"), y_pred.astype("int32")))

print("RECALL : ", recall_score(y_test.astype("int32"), y_pred.astype("int32")))

print("PRECISION : ", precision_score(y_test.astype("int32"), y_pred.astype("int32")))

print("F1 : ", f1_score(y_test.astype("int32"), y_pred.astype("int32")))

confusion_matrix(y_test.astype("int32"), y_pred.astype("int32"))
```

# DEEP NEURAL NETWORK

### Model Building


```python
model2 = Sequential()

model2.add(Dense(12, input_dim=13, activation='relu'))

model2.add(Dense(8, activation='relu'))

model2.add(Dense(1, activation='sigmoid'))
```


### Model Summary

```python
model2.summary()
```


### Model Training

```python
model2.compile(optimizer = SGD(learning_rate=0.01, momentum = 0.01), loss="binary_crossentropy", metrics = ['accuracy'])
```

```python
model2.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test), batch_size=len(X_train))
```

```python
y_pred2 = model2.predict(X_test)
y_pred2 = (y_pred2 > 0.5 )
```


### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,  f1_score, precision_score

print("ACCURACY : ", accuracy_score(y_test.astype("int32"), y_pred2.astype("int32")))

print("RECALL : ", recall_score(y_test.astype("int32"), y_pred2.astype("int32")))

print("PRECISION : ", precision_score(y_test.astype("int32"), y_pred2.astype("int32")))

print("F1 : ", f1_score(y_test.astype("int32"), y_pred2.astype("int32")))

confusion_matrix(y_test.astype("int32"), y_pred2.astype("int32"))
```
