### Blending of Images

```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

```python
image1 = cv2.imread("D:\\Computer Vision\\data\\windows1.jpg")
image2 = cv2.imread("D:\\Computer Vision\\data\\apple.png")

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

print(image1.shape)
print(image2.shape)
```

- Resizing the Image because both should have same dimensions for blending

```python
image1 = cv2.resize(image1, (478, 478))
image2 = cv2.resize(image2, (478, 478))
```

```python
plt.subplot(1,2,1)
plt.axis("off")
plt.imshow(image1)

plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(image2)
```

- Blending images

```python
blended_image1 = cv2.addWeighted(src1=image1, alpha=0.6, src2=image2, beta=0.2, gamma=0)
blended_image2 = cv2.addWeighted(src1=image1, alpha=0.5, src2=image2, beta=0.5, gamma=0)
blended_image3 = cv2.addWeighted(src1=image1, alpha=0.2, src2=image2, beta=0.6, gamma=0)

plt.subplot(1,3,1)
plt.axis("off")
plt.imshow(blended_image1)

plt.subplot(1,3,2)
plt.axis("off")
plt.imshow(blended_image2)

plt.subplot(1,3,3)
plt.axis("off")
plt.imshow(blended_image3)
```
