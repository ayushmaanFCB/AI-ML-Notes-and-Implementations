### Memory Addresses

```python
from PIL import Image
import matplotlib.pyplot as plt, numpy
import cv2
```

```python
image = cv2.imread("D:\\Computer Vision\\data\\weeknd.jpeg")

print(type(image))
id(image)               # Memory Address
```

- Direct assignment: refers to same memory address

```python
copy1 = image
id(copy1)
```

- Using copy(): different memory locations assigned

```python
copy2 = image.copy()
id(copy2)
```

On making the original image pixels to 0, the copy1 value gets affected but copy2 is independent in a different memory location.

```python
image[:, :, ] = 0

plt.subplot(1, 2, 1)
plt.imshow(copy1)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(copy2)
plt.axis('off')
```

### Image Flipping

```python
from PIL import ImageOps

image = Image.open("D:\\Computer Vision\\data\\weeknd.jpeg")
```

```python
image_flip1 = ImageOps.flip(image)          # VERTICAL

image_flip2 = ImageOps.mirror(image)        # HORIZONTAL
```

```python
plt.subplot(1, 3, 1)
plt.axis('off')
plt.imshow(image)

plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(image_flip1)

plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(image_flip2)
```
