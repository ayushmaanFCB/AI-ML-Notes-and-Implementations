# Edge Detection

```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('D:\\CSE 410 - Computer Vision\\data\\weeknd.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,3,1)
plt.imshow(img), plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_rgb), plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img_gray, cmap="gray"), plt.axis("off")
```

### Invoke Canny Edge Detection

- `threshold1=127` and `threshold2=127`: These are the lower and upper thresholds for edge detection. Pixels with gradient values below threshold1 are considered non-edges, and those above threshold2 are definitely edges. Pixels with gradient values between these thresholds are considered edges if they are connected to strong edges.

The Canny edge detection algorithm works by:
- Calculating the gradient magnitude of the image.
- Suppressing weak edges.
- Linking strong edges to form continuous contours.


```python
edges = cv2.Canny(image=img, threshold1=180, threshold2=227)

plt.imshow(edges, cmap="gray")
```

### To Reduce Noise

- Blurring
- Lower threhold is maximum of 0 and 70% of median value
- Upper threshold is minimum of 255 and 0.13% of median value
- Apply canny detection

```python
blurred = cv2.blur(img_gray, ksize = (5, 5))
med_val = np.median(blurred)

lower = int(max(0, 0.7 * med_val))
upper = int(min(255, 1.3 * med_val))
print(lower, upper)

edges = cv2.Canny(image=img_gray, threshold1=lower, threshold2=upper)
plt.imshow(edges, cmap='gray'), plt.axis('off')
```

# Grid Detection

```python
import numpy as np, matplotlib.pyplot as plt
import cv2

img = cv2.imread('D:\\CSE 410 - Computer Vision\\data\\flat_chessboard.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img), plt.axis("off")
```

### Invoke findChessboardCorners

- `found`: This variable will hold a boolean value indicating whether the corners were successfully found.
- `corners`: If found is True, this variable will store the pixel coordinates of the detected corners.


```python
found, corners = cv2.findChessboardCorners(img, (7,7))

if found:
    print("Open CV was able to find the corners")
else:
    print("Open CV was not able to find the corners")
```

```python
print(corners.shape)
```

### Invoke drawChessboardCorners

```python
chess1_corners = cv2.drawChessboardCorners(img, patternSize=(7,7), corners=corners, patternWasFound=found)

plt.imshow(chess1_corners), plt.axis("off")
```
