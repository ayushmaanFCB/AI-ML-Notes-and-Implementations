# Morphological Operations

```python
import numpy as np, pandas as pandas
import cv2
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='vision', org=(50, 300), thickness=30,
        fontFace=font, fontScale=5, color=(255, 255, 255), lineType=cv2.LINE_AA
    )
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    plt.axis("off")
```

```python
img = load_img()
display_img(img)
```

### Creating the Kernel

```python
kernel = np.ones((5, 5), np.uint8)
```

### Erosion

- shrinks the boundaries of an object in an image

```python
erosion1 = cv2.erode(img, kernel, iterations = 3)
display_img(erosion1)
```

```python
erosion2 = cv2.erode(img, kernel, iterations = 5)
display_img(erosion2)
```

### Dilation

```python
img = load_img()

dilated1 = cv2.dilate(img, kernel, iterations=2)
display_img(dilated1)
```

```python
dilated2 = cv2.dilate(img, kernel, iterations=10)
display_img(dilated2)
```

### Opening

```python
img = load_img()
```

- Generating a 600x600 matrix where each element is random integer between 0(inclusive) and 2(exclusive)

```python
white_noise = np.random.randint(low=0, high=2, size=(600,600))

display_img(white_noise)
```

- Adding noise to image

***NOTE***: Multiply white noise with 255, as original image is a grayscale image but white nowise only have 0s and 1s.

```python
noised_img = img + (white_noise * 255)
display_img(noised_img)
```

- OPENING

```python
opening1 = cv2.morphologyEx(noised_img, cv2.MORPH_OPEN, kernel)
display_img(opening1)
```

### Closing

```python
img = load_img()
```

- Generating black noise

```python
black_noise = np.random.randint(low=0, high=2, size=(600,600))

display_img(black_noise)
```

- Adding noise to image

```python
noised_img2 = img + (black_noise * (-255))
noised_img2[noised_img2 == -255] = 0
display_img(noised_img2)
```

- CLOSING

```python
closing1 = cv2.morphologyEx(noised_img2, cv2.MORPH_CLOSE, kernel)
display_img(closing1)
```

### Morphological Gradient

```python
img = load_img()
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

display_img(gradient)
```

### Tophat

```python
img = load_img()

tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

display_img(tophat)
```

```python
img = load_img()

blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

display_img(blackhat)
```
