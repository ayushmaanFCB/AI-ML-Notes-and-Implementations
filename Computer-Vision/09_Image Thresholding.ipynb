# Thresholding

- Thresholding is a fundamental technique in computer vision used to convert a grayscale image into a binary image. This process simplifies the image, making it easier to analyze and identify regions of interest. Hereâ€™s how it works:
    - Grayscale Image: Thresholding typically starts with a grayscale image, where each pixel has an intensity value between 0 (black) and 255 (white).
    - Threshold Value: A specific threshold value is chosen. Pixels with intensity values above this threshold are set to white (255), and those below are set to black (0)

# Global Thresholding

```python
import cv2
import numpy as np, matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("D:\\CSE 410 - Computer Vision\\data\\weeknd.jpeg", 0)      
# 0: indicates reading as a grayscale image

plt.axis("off"), plt.imshow(img1, cmap='gray'), plt.show()

print(img1.shape)           # No 3rd dimension as Grayscale
```

### 1. Binary Thresholding

```python
threshold1, ret1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
# (source, threshold, maxVal, type)

print(threshold1)

plt.axis("off"), plt.imshow(ret1, cmap='gray'), plt.show()
```

### 2. Binary Thresholding Inverted

```python
threshold2, ret2 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY_INV)
# (source, threshold, maxVal, type)

print(threshold2)

plt.axis("off"), plt.imshow(ret2, cmap='gray'), plt.show()
```

### 3. Truncated Thresholding

```python
threshold3, ret3 = cv2.threshold(img1, 127, 255, cv2.THRESH_TRUNC)
# (source, threshold, maxVal, type)

print(threshold3)

plt.axis("off"), plt.imshow(ret3, cmap='gray'), plt.show()
```

### 4. Threshold to Zero

```python
threshold4, ret4 = cv2.threshold(img1, 127, 255, cv2.THRESH_TOZERO)
# (source, threshold, maxVal, type)

print(threshold4)

plt.axis("off"), plt.imshow(ret4, cmap='gray'), plt.show()
```

### 5. Threshold to Zero Inverted

```python
threshold5, ret5 = cv2.threshold(img1, 127, 255, cv2.THRESH_TOZERO_INV)
# (source, threshold, maxVal, type)

print(threshold5)

plt.axis("off"), plt.imshow(ret5, cmap='gray'), plt.show()
```

Comparisons

```python
outputs = [img1, ret1, ret2, ret3, ret4, ret5]

plt.figure(figsize=(12,8))

for i in range(1,7):
    plt.subplot(1,6,i)
    plt.axis("off")
    plt.imshow(outputs[i-1], cmap='gray')

plt.show()
```

# Adaptive Thresholding

```python
img2 = cv2.imread("D:\\CSE 410 - Computer Vision\\data\\crossword.jpg", 0)      
# 0: indicates reading as a grayscale image

plt.axis("off"), plt.imshow(img2, cmap='gray'), plt.show()

print(img2.shape)           # No 3rd dimension as Grayscale
```

### 1. Mean Adaptive Thresholding

- (source, maxVal, type, threshold_of_neighborhood, neighborhood_dimension, constant_C)

```python
th_img = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11 , 8)

plt.figure(figsize=(7,7))
plt.axis("off"), plt.imshow(th_img, cmap="gray"), plt.show()
```

### 2. Gaussian Adaptive Thresholding

```python
th_img2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11 , 8)

plt.figure(figsize=(7,7))
plt.axis("off"), plt.imshow(th_img2, cmap="gray"), plt.show()
```

<hr>

### Task (blending)

```python
blended_img = cv2.addWeighted(th_img, 0.7, th_img2, 0.3, 0.0,)

plt.figure(figsize=(7,7))
plt.axis("off"), plt.imshow(blended_img, cmap="gray"), plt.show()
```
