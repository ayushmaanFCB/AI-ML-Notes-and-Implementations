### Introduction to OpenCV

```python
# !pip install opencv-python

import cv2
import matplotlib.pyplot as plt, numpy
```

- Reading an image

Here we won't get the exact colors on reading because PIL reads image in the order (R,G,B) but OpenCV reads in order (B,G,R)

```python
image = cv2.imread("D:\\Computer Vision\\data\\metro_future.jpeg")

print(type(image))

plt.axis('off')
plt.imshow(image)
```

```python
print(image.shape)
```

```python
print(image[0])         # PIL would display image directly, her it shows intensity values
```

```python
print(image.min())

print(image.max())
```

### Color Conversion

- To Coloured RGB Image

```python
converted_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(converted_image)
```

- To Grayscale Image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(gray_image, cmap='gray')
```

- Reading image directly as a Grayscale

```python
image3 = cv2.imread("D:\\Computer Vision\\data\\metro_future.jpeg", cv2.IMREAD_GRAYSCALE)

plt.axis('off')
plt.imshow(image3, cmap='gray')
```

### Saving images locally

```python
cv2.imwrite("grey_future_metro.jpg", image3)
```

### Separating color

```python
# here, order is (B,G,R)

blue, green, red = image[:, :, 0], image[:, :, 1], image[:, :, 2]
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("BLUE")
plt.imshow(blue, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("GREEN")
plt.imshow(green, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("RED")
plt.imshow(red, cmap='gray')
plt.axis('off')
```
