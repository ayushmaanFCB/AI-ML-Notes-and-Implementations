# Contours Detection

- Line or curve that joins all the points along the boundary of an object in an image that have the same intensity or color. 
- Contour is a continuous line or curve that defines the complete boundary of an object. 

```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('D:\\CSE 410 - Computer Vision\\data\\weeknd.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,3,1)
plt.imshow(img), plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_rgb), plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img_gray, cmap="gray"), plt.axis("off")
```
