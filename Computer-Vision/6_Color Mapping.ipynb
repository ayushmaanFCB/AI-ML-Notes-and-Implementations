### Color Mapping

```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("D:\\Computer Vision\\data\\weeknd.jpeg")
plt.axis("off")
plt.imshow(image)
```

```python
image1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(image1)
```

- HSVL: Hue, Saturation, and Value

```python
image2 = cv2.cvtColor(image1, cv2.COLOR_RGB2HSV)

plt.axis("off")
plt.imshow(image2)
```

```python
image3 = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.axis("off")
plt.imshow(image3)
```

- HLS: Hue, Lightness, Saturation

```python
image4 = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)

plt.axis("off")
plt.imshow(image4)
```
