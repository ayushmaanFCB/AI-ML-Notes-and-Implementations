### Flipping using CV2

```python
import cv2

image_cv = cv2.imread('bat1.jpeg')
image_cv = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)
```

- Vertical Flipping

```python
flipped_vert = cv2.flip(image_cv, 0)        # 0 -> Vertical Flip

plt.axis('off')
plt.imshow(flipped_vert)
```

- Horizontal Flipping

```python
flipped = cv2.flip(image_cv, 1)        # 1 -> Horizontal Flip

plt.axis('off')
plt.imshow(flipped)
```

- Flipping around both the axes

```python
flipped = cv2.flip(image_cv, -1)

plt.axis('off')
plt.imshow(flipped)
```

## Display Image as Large Size

```python
plt.figure(figsize=(15,8))
plt.axis('off')
plt.imshow(image_cv)
```

### Resizing the Image

Interpolations:
- INTER_NEAREST
- INTER_LINEAR
- INTER_CUBIC
- INTER_LANCZOS4
- INTER_AREA


```python
resized_img = cv2.resize(image_cv, (75,100), interpolation=cv2.INTER_NEAREST)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```

```python
resized_img = cv2.resize(image_cv, (50, 25), interpolation=cv2.INTER_LINEAR)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```

```python
resized_img = cv2.resize(image_cv, (25,50), interpolation=cv2.INTER_CUBIC)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```

```python
resized_img = cv2.resize(image_cv, (100,100), interpolation=cv2.INTER_LANCZOS4)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```

```python
resized_img = cv2.resize(image_cv, (30,15), interpolation=cv2.INTER_AREA)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```

```python
cv2.imwrite('resized_img.jpg', resized_img)
```

- We can pass two more parameters `fx` and `fy`, in case we don't pass the output dimension, they will be the scaling factors.

```python
resized_img = cv2.resize(image_cv, (0,0), fx=0.25, fy=0.15, interpolation=cv2.INTER_LINEAR)

print(resized_img.shape)

plt.axis('off')
plt.imshow(resized_img)
```
