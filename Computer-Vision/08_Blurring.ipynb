### Loading Image and Pre-Processing

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
````

````python
def load_img(path="D:\\CSE 410 - Computer Vision\\data\\weeknd.jpeg"):
    img = cv2.imread(path).astype(np.float32) / 255
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
def display_img(img):
    fig = plt.figure(figsize=(8,3))
    ax = fig.add_subplot(111)
    ax.imshow(img)
````

````python
img1 = load_img()
display_img(img1)
````

### Gamma Correction for adjusting Brightness

- Gamma can't be 0 or negative

````python
gamma = 0.4

gamma_image1 = np.power(img1, gamma)
display_img(gamma_image1)
````

````python
gamma = 0.1

gamma_image2 = np.power(img1, gamma)
display_img(gamma_image2)
````

### Blurring

- Blurring with custom kernel

    - The ```cv2.filter2D()``` function in OpenCV is used to apply a custom filter to an image.

    - ```ddepth```: This specifies the depth of the output image. A value of -1 means the output image will have the same depth as the source image.

````python
kernel_ = np.ones(shape=(5,5), dtype=np.float32) / 12
kernel_
````

````python
blurred_img = cv2.filter2D(img1, kernel=kernel_, ddepth=-1)
plt.imshow(blurred_img)
````

- Mean Blur

````python
blurred_img = cv2.blur(img1, (5,5))
display_img(blurred_img)
````

- Gaussian Blur

    - ```sigmaX```: The standard deviation in the X direction
    - ```sigmaY```: The standard deviation in the Y direction

````python
blurred_img = cv2.GaussianBlur(img1, (5,5), sigmaX=100, sigmaY=20)
display_img(blurred_img)
````

- Median Blur

````python
blurred_img = cv2.medianBlur(img1, 5, 100)
display_img(blurred_img)
````
