```python
from PIL import ImageOps

img_grey = ImageOps.grayscale(img)

plt.imshow(img_grey, cmap="gray")           # by default matplotlib has different cmap
```

### Color Images

```python
img = Image.open("D:\\Computer Vision\\data\\metro_future.jpeg")

red, green, blue = img.split()
```

Each red, blue and green will now be a greyscale representation only with varying intensities

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(red, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(blue, cmap='gray')
plt.axis('off')
```

- Converting from PIL to Numpy

```python
import numpy 

img_arr = numpy.array(img)

print(img_arr[0])
```

We need to use Numpy in conjunction with PIL

```python
parrot = Image.open("D:\\Computer Vision\\data\\parrot.jpeg")

parrot_arr = numpy.asarray(parrot)          # or numpy.array()

plt.imshow(parrot)
plt.axis('off')
```

```python
print(type(parrot_arr))

print(parrot_arr.shape)
```

- Extracting channels using slicing

```python
pic_red = parrot_arr.copy()[:, :, 0]
pic_green = parrot_arr.copy()[:, :, 1]
pic_blue = parrot_arr.copy()[:, :, 2]

print(pic_red.shape)

plt.figure(figsize=(17,5))

plt.subplot(1, 3, 1)
plt.imshow(pic_red, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(pic_green, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(pic_blue, cmap='gray')
plt.axis('off')
```

```python
test = parrot_arr.copy()

test[:, :, 0] = 0
test[:, :, 1] = 0
test[:, :, 2] = 0

plt.imshow(test)
plt.axis('off')
```

### Creating a new Grayscale Image

```python
width, height = 256, 100

image = Image.new('L', (width, height))         # 100 rows and 256 columns

plt.imshow
```

```python
for x in range(width):
    for y in range(height):
        image.putpixel((x, y), x)

image.save("gradient.png")

plt.axis('off')
plt.imshow(image, cmap='gray')
```

### Creating Color Images

```python
width, height = 100, 100
color = (255, 0, 0)             # RED color

image = Image.new('RGB', (width, height), color=color)

plt.axis('off')
plt.imshow(image)
```
