### Creating a Blank Image

```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
plt.axis('off')
plt.imshow(blank_img)
```

### Drawing Rectangle

```python
shaped1 = cv2.rectangle(blank_img, pt1=(100, 100), pt2=(200, 200), color=(0, 255, 0), thickness=5)
plt.axis('off')
plt.imshow(shaped1)
```

```python
cv2.imwrite('rectangle.jpg', shaped1)
```

- Filled rectangle

```python
shaped1 = cv2.rectangle(blank_img, pt1=(250, 250), pt2=(310, 310), color=(0, 255, 0), thickness=-1)
plt.axis('off')
plt.imshow(shaped1)
```

### Drawing Circle

```python
shaped2 = cv2.circle(blank_img, center=(100, 450), radius=50, color=(0, 255, 0), thickness=5)
plt.axis('off')
plt.imshow(shaped2)
```

```python
shaped2 = cv2.circle(blank_img, center=(440, 440), radius=50, color=(0, 255, 0), thickness=-1)
plt.axis('off')
plt.imshow(shaped2)
```

### Drawing Line

```python
shaped1 = cv2.line(blank_img, pt1=(450, 450), pt2=(100, 100), color=(0, 255, 0), thickness=5)
plt.axis('off')
plt.imshow(shaped1)
```

### Drawing Text

```python
shaped2 = cv2.putText(blank_img, text="Hello", org=(250, 250),
                      fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1,
                      color=(255, 255, 0), thickness=2,
                      lineType = cv2.LINE_AA)

plt.axis('off')
plt.imshow(shaped2)
```

```python
shaped2 = cv2.putText(blank_img, text="Hello", org=(250, 250),
                      fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1,
                      color=(255, 0, 0), thickness=2,
                      bottomLeftOrigin = True)          # INVERTED TEXT

plt.axis('off')
plt.imshow(shaped2)
```

### Drawing Polygon

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
height, width = blank_img.shape[:2]

changed_height, changed_width = height//2, width//2

print(changed_height, " ", changed_width)
```

```python
pts = np.array([[0, changed_height], [width, changed_height], [changed_width, 0], [changed_width, height]])

cv2.polylines(blank_img, [pts], isClosed=True, color=(0,255,0), thickness=10)

plt.axis("off")
plt.imshow(blank_img)
```

### Cropping an Image

```python
import cv2, matplotlib.pyplot as plt, numpy

image = cv2.imread("D:\\Computer Vision\\data\\weeknd.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print(image.shape)

plt.axis("off")
plt.imshow(image)
```

```python
x, y, w, h = 50, 50, 100, 100

# cropping from top left corner
cropped = image[y:y+h, x:x+w]           # startY:endY, startX:endX

plt.axis("off")
plt.imshow(cropped)
```

### Assignment : 

Write a script that draws a blue circle with a radius of 50 pixes in the center of the image *"example.jpg"* and saves it as *"circled_example.jpg"*.

```python
image = cv2.imread("D:\\Computer Vision\\data\\metro_weeknd.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print(image.shape)

plt.axis("off")
plt.imshow(image)
```

```python
center_point = (image.shape[1]//2, image.shape[0]//2)
center_point
```

```python
shaped2 = cv2.circle(image, center=center_point, radius=50, color=(0, 0, 255), thickness=20)

cv2.imwrite("circled_example.jpg", shaped2)

plt.axis('off')
plt.imshow(shaped2)
```
